```python
# my implementation

def parse_conversation(text: str):   # given my Anurudh 
    # Split into lines and strip whitespace
    import re
    lines = [line.strip() for line in text.strip().splitlines() if line.strip()]
    
    conversation = []
    for line in lines:
        match = re.match(r"^(U|A):\s*'(.*)'$", line)
        if match:
            role = "user" if match.group(1) == "U" else "assistant"
            message = match.group(2)
            conversation.append({"role": role, "content": message})
    
    return conversation


def clean_message(msg: str):   # cleans message and removes leadning and trailing stray quotes - required for telugu
         return msg.strip().strip("'‘’\"").strip()

def translate_to_english(text: str, auto_translate = True):  # translated to english it required - malfunctions sometimes, changes the script to english but looks like telugu written in english
        from googletrans import Translator

        translator = Translator()
        if not auto_translate:
            return text
        try:
            detection = translator.detect(text)
            if detection.lang != "en":
                return translator.translate(text, src=detection.lang, dest="en").text
        except Exception:
            return text  # fallback if translation fails
        return text


def parse_conversation_seg(text: str): # modified parser - takes care of unlabelled lines, appends them into the previous token and maintains a buffer of stray sentences to ab appended at the end
    import re              # saves all data in pairs of User--Assistant, removes the 1st entry if its from assistant and returns list of dictionary pairs conataining User-assiatnt paired messages
    lines = [line.strip() for line in text.strip().splitlines() if line.strip()]
    
    conversation = []
    current_pair = {}
    token = None
    buffer = []
    role = None

    pattern = re.compile(r"^(U|A):\s*['‘’\"]?(.*)")

    for line in lines:
        line = translate_to_english(line)
        match = pattern.match(line)

        if match:
            if role and buffer:
                content = clean_message(" ".join(buffer))
                if role == 'User':
                    if 'User' in current_pair:
                        current_pair['User'] = current_pair['User'] + " " + content
                    else:
                        current_pair['User'] = content
                else:
                    if 'Assistant' in current_pair:
                        current_pair['Assistant'] = current_pair['Assistant'] + " " + content
                    else:
                        current_pair['Assistant'] = content
                buffer = []

            if "User" in current_pair and "Assistant" in current_pair:
                conversation.append(current_pair)
                current_pair = {}
            if match.group(1) == 'U':
                role = 'User'
            else:
                role = 'Assistant'
            
            content = clean_message(match.group(2))

            if not conversation and not current_pair and role == "Assistant":
                continue
            buffer = [content]
        else:
            buffer.append(clean_message(line))

    if role and buffer:
        content = clean_message(" ".join(buffer))
        if role == 'User':
            if 'User' in current_pair:
                current_pair['User'] = current_pair['User'] + " " + content
            else:
                current_pair['User'] = content
        else:
            if 'Assistant' in current_pair:
                current_pair['Assistant'] = current_pair['Assistant'] + " " + content
            else:
                current_pair['Assistant'] = content

    if current_pair:
        conversation.append(current_pair)

    return conversation


def conversation_modifier(conversation, token): # modifies the original made paired dictionary - if you only want to use 'user' or only 'assistant' for segmentation
    if token is None:
        return conversation
    modified_conversation = []
    for conv in conversation:
          if token in conv:
            modified_conversation.append({token: conv[token]})
    return modified_conversation
```

```python
# simple implemnetation using cosine distance - good for vectorized embeddings

def cosine(a, b): # simple cosine distance between pairs of vectors
        import numpy as np
        return sum(x*y for x, y in zip(a, b)) / ((np.sqrt(sum(x*x for x in a)) or 1.0) * (np.sqrt(sum(y*y for y in b)) or 1.0))

def gen_episodes(conversation, cosine_thresh): # generate episodes using embedding from a hugging face LLM and coisne distance
    from sentence_transformers import SentenceTransformer
    import numpy as np

    model = SentenceTransformer("all-MiniLM-L6-v2") # hugging face LLM which converts sentences in a 384 length long vector 
    txt_2_embed = []
    if all('User' in pair and 'Assistant'in pair for pair in conversation):
        for pair in conversation:
            txt_2_embed.append(pair["User"] + " " + pair["Assistant"])
    elif all("User" in pair for pair in conversation):
        for pair in conversation:
            txt_2_embed.append(pair["User"])
    elif all("Assistant" in pair for pair in conversation):
        for pair in conversation:
            txt_2_embed.append(pair["Assistant"]) 

    embeddings = model.encode(txt_2_embed)

    episodes = []
    current_episode = [conversation[0]]
    scores = []
    for ii in range(len(conversation)-1):
        scr = cosine(embeddings[ii], embeddings[ii+1])
        scores.append(scr)
        if scr < cosine_thresh:
             episodes.append(current_episode)
             current_episode = [conversation[ii+1]]
        else:
             current_episode.append(conversation[ii+1])
    if current_episode:
         episodes.append(current_episode)
    return episodes, scores
```

```python
# surprise based computation - computationally heavy and not working good, implemented paper : https://arxiv.org/abs/2407.09450
import torch 
from transformers import AutoTokenizer, AutoModelForCausalLM
import numpy as np

tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
model = AutoModelForCausalLM.from_pretrained("distilgpt2")
model.eval()

def compute_surprise(text): # compute surprise using tokenization of texts and finidng the negatove log likelihood of a text given a context (series of previous texts)
    input = tokenizer(text, return_tensors='pt')
    with torch.no_grad():
        output = model(**input, labels=input['input_ids'])
        return output.loss.item()
    
def surprise_episodes(conversation, gamma, window): # calcualtes a dynamic threhold of liklihood in a moving window fashion. starts a new episode when the likelihood of a word crosses threhold
    segment = []
    if all('User' in pair and 'Assistant'in pair for pair in conversation):
        for pair in conversation:
            segment.append(pair["User"] + " " + pair["Assistant"])
    elif all("User" in pair for pair in conversation):
        for pair in conversation:
            segment.append(pair["User"])
    elif all("Assistant" in pair for pair in conversation):
        for pair in conversation:
            segment.append(pair["Assistant"]) 

    surprise_idx =[]
    for ss in segment:
        sr = compute_surprise(ss)
        surprise_idx.append(sr)
    
    episodes = []
    current_episode = [conversation[0]]

    for ii in range(1, len(conversation)): # takes a sequence of words in a moving window
        win = surprise_idx[max(0, ii-window):ii]
        if win: 
            mn = np.mean(win)
            std = np.std(win)
        else:
            mn = 0
            std = 1
        thresh = mn + gamma*std  # calculates dynamic threshold

        if surprise_idx[ii] > thresh:
            episodes.append(current_episode)
            current_episode = [conversation[ii]]
        else:
            current_episode.append(conversation[ii])

        
    if current_episode:
        episodes.append(current_episode)

    return episodes, surprise_idx
```
